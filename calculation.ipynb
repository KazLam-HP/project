```python
import pandas as pd
import plotly.express as px

# Load and process the data for origin
df_orig = pd.read_csv("data/ref_country_orig.csv")
df_orig = df_orig.melt(id_vars=["country"], var_name="Year", value_name="Refugees_origin")
df_orig = df_orig.dropna(subset=["Refugees_origin"])
df_orig["Year"] = df_orig["Year"].astype(int)

# Load and process the data for asylum
df_asylm = pd.read_csv("data/ref_country_asylm.csv")
df_asylm = df_asylm.melt(id_vars=["country"], var_name="Year", value_name="Refugees_asylum")
df_asylm = df_asylm.dropna(subset=["Refugees_asylum"])
df_asylm["Year"] = df_asylm["Year"].astype(int)

# Merge datasets
df_combined = pd.merge(df_orig, df_asylm, on=["country", "Year"], how="inner")

# Global totals
df_total = df_combined.groupby("Year").agg({
    "Refugees_origin": "sum",
    "Refugees_asylum": "sum"
}).reset_index()

df_total["Total"] = df_total["Refugees_origin"] + df_total["Refugees_asylum"]

import pandas as pd
import plotly.express as px

# Load and process the data for origin
df_orig = pd.read_csv("data/ref_country_orig.csv")
df_orig = df_orig.melt(id_vars=["country"], var_name="Year", value_name="Refugees_origin")
df_orig = df_orig.dropna(subset=["Refugees_origin"])
df_orig["Year"] = df_orig["Year"].astype(int)

# Load and process the data for asylum
df_asylm = pd.read_csv("data/ref_country_asylm.csv")
df_asylm = df_asylm.melt(id_vars=["country"], var_name="Year", value_name="Refugees_asylum")
df_asylm = df_asylm.dropna(subset=["Refugees_asylum"])
df_asylm["Year"] = df_asylm["Year"].astype(int)

# Merge datasets
df_combined = pd.merge(df_orig, df_asylm, on=["country", "Year"], how="inner")

# Global totals
df_total = df_combined.groupby("Year").agg({
    "Refugees_origin": "sum",
    "Refugees_asylum": "sum"
}).reset_index()

df_total["Total"] = df_total["Refugees_origin"] + df_total["Refugees_asylum"]

import pandas as pd
import plotly.express as px

# Load and process the data for origin
df_orig = pd.read_csv("data/ref_country_orig.csv")
df_orig = df_orig.melt(id_vars=["country"], var_name="Year", value_name="Refugees_origin")
df_orig = df_orig.dropna(subset=["Refugees_origin"])
df_orig["Year"] = df_orig["Year"].astype(int)

# Load and process the data for asylum
df_asylm = pd.read_csv("data/ref_country_asylm.csv")
df_asylm = df_asylm.melt(id_vars=["country"], var_name="Year", value_name="Refugees_asylum")
df_asylm = df_asylm.dropna(subset=["Refugees_asylum"])
df_asylm["Year"] = df_asylm["Year"].astype(int)

# Merge datasets
df_combined = pd.merge(df_orig, df_asylm, on=["country", "Year"], how="inner")

# Global totals
df_total = df_combined.groupby("Year").agg({
    "Refugees_origin": "sum",
    "Refugees_asylum": "sum"
}).reset_index()

df_total["Total"] = df_total["Refugees_origin"] + df_total["Refugees_asylum"]

import pandas as pd
import plotly.express as px

# Load and process the data for origin
df_orig = pd.read_csv("data/ref_country_orig.csv")
df_orig = df_orig.melt(id_vars=["country"], var_name="Year", value_name="Refugees_origin")
df_orig = df_orig.dropna(subset=["Refugees_origin"])
df_orig["Year"] = df_orig["Year"].astype(int)

# Load and process the data for asylum
df_asylm = pd.read_csv("data/ref_country_asylm.csv")
df_asylm = df_asylm.melt(id_vars=["country"], var_name="Year", value_name="Refugees_asylum")
df_asylm = df_asylm.dropna(subset=["Refugees_asylum"])
df_asylm["Year"] = df_asylm["Year"].astype(int)

# Merge datasets
df_combined = pd.merge(df_orig, df_asylm, on=["country", "Year"], how="inner")

# Global totals
df_total = df_combined.groupby("Year").agg({
    "Refugees_origin": "sum",
    "Refugees_asylum": "sum"
}).reset_index()

df_total["Total"] = df_total["Refugees_origin"] + df_total["Refugees_asylum"]

import pandas as pd
import plotly.express as px

# Load and process the data for origin
df_orig = pd.read_csv("data/ref_country_orig.csv")
df_orig = df_orig.melt(id_vars=["country"], var_name="Year", value_name="Refugees_origin")
df_orig = df_orig.dropna(subset=["Refugees_origin"])
df_orig["Year"] = df_orig["Year"].astype(int)

# Load and process the data for asylum
df_asylm = pd.read_csv("data/ref_country_asylm.csv")
df_asylm = df_asylm.melt(id_vars=["country"], var_name="Year", value_name="Refugees_asylum")
df_asylm = df_asylm.dropna(subset=["Refugees_asylum"])
df_asylm["Year"] = df_asylm["Year"].astype(int)

# Merge datasets
df_combined = pd.merge(df_orig, df_asylm, on=["country", "Year"], how="inner")

# Global totals
df_total = df_combined.groupby("Year").agg({
    "Refugees_origin": "sum",
    "Refugees_asylum": "sum"
}).reset_index()

df_total["Total"] = df_total["Refugees_origin"] + df_total["Refugees_asylum"]

import pandas as pd
import plotly.express as px

# Load and process the data for origin
df_orig = pd.read_csv("data/ref_country_orig.csv")
df_orig = df_orig.melt(id_vars=["country"], var_name="Year", value_name="Refugees_origin")
df_orig = df_orig.dropna(subset=["Refugees_origin"])
df_orig["Year"] = df_orig["Year"].astype(int)

# Load and process the data for asylum
df_asylm = pd.read_csv("data/ref_country_asylm.csv")
df_asylm = df_asylm.melt(id_vars=["country"], var_name="Year", value_name="Refugees_asylum")
df_asylm = df_asylm.dropna(subset=["Refugees_asylum"])
df_asylm["Year"] = df_asylm["Year"].astype(int)

# Merge datasets
df_combined = pd.merge(df_orig, df_asylm, on=["country", "Year"], how="inner")

# Global totals
df_total = df_combined.groupby("Year").agg({
    "Refugees_origin": "sum",
    "Refugees_asylum": "sum"
}).reset_index()

df_total["Total"] = df_total["Refugees_origin"] + df_total["Refugees_asylum"]

import pandas as pd
import plotly.express as px

# Load and process the data for origin
df_orig = pd.read_csv("data/ref_country_orig.csv")
df_orig = df_orig.melt(id_vars=["country"], var_name="Year", value_name="Refugees_origin")
df_orig = df_orig.dropna(subset=["Refugees_origin"])
df_orig["Year"] = df_orig["Year"].astype(int)

# Load and process the data for asylum
df_asylm = pd.read_csv("data/ref_country_asylm.csv")
df_asylm = df_asylm.melt(id_vars=["country"], var_name="Year", value_name="Refugees_asylum")
df_asylm = df_asylm.dropna(subset=["Refugees_asylum"])
df_asylm["Year"] = df_asylm["Year"].astype(int)

# Merge datasets
df_combined = pd.merge(df_orig, df_asylm, on=["country", "Year"], how="inner")

# Global totals
df_total = df_combined.groupby("Year").agg({
    "Refugees_origin": "sum",
    "Refugees_asylum": "sum"
}).reset_index()

df_total["Total"] = df_total["Refugees_origin"] + df_total["Refugees_asylum"]

import pandas as pd
import plotly.express as px

# Load and process the data for origin
df_orig = pd.read_csv("data/ref_country_orig.csv")
df_orig = df_orig.melt(id_vars=["country"], var_name="Year", value_name="Refugees_origin")
df_orig = df_orig.dropna(subset=["Refugees_origin"])
df_orig["Year"] = df_orig["Year"].astype(int)

# Load and process the data for asylum
df_asylm = pd.read_csv("data/ref_country_asylm.csv")
df_asylm = df_asylm.melt(id_vars=["country"], var_name="Year", value_name="Refugees_asylum")
df_asylm = df_asylm.dropna(subset=["Refugees_asylum"])
df_asylm["Year"] = df_asylm["Year"].astype(int)

# Merge datasets
df_combined = pd.merge(df_orig, df_asylm, on=["country", "Year"], how="inner")

# Global totals
df_total = df_combined.groupby("Year").agg({
    "Refugees_origin": "sum",
    "Refugees_asylum": "sum"
}).reset_index()

df_total["Total"] = df_total["Refugees_origin"] + df_total["Refugees_asylum"]
```
